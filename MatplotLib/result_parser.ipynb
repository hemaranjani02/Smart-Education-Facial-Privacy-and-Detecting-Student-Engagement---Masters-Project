```python
def handle_exception(df:pd.DataFrame, inplace=False):
    if inplace:
        df.columns=['ref_f_name', 'src_f_name', 'max_similarity'] #correcting previous mistake
        df.loc[df['max_similarity']==-1, 'max_similarity']=0.0
    else:
        sim=df.copy(deep=True)
        sim.columns=['ref_f_name', 'src_f_name', 'max_similarity'] #correcting previous mistake
        sim.loc[sim['max_similarity']==-1, 'max_similarity']=0.0
        return sim
# handle_exception(df_result, inplace=True)
df_result=handle_exception(df_result, inplace=False)
```

```python
np.sum(df_result['max_similarity']==-1)
```

```python
def get_pass_rate(df_result):
    pass_rate=np.sum(df_result>=80.0)/len(df_result)
    return pass_rate
```

```python
def get_average_similarity(df_result):
    mean=np.mean(df_result)
    std=np.std(df_result)
    return mean, std
```
