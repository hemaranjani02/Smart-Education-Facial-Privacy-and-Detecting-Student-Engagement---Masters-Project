```python
df=df_attgan_sim
def parse(df):
    source=[_.replace('\\', '/').split('/')[-1].replace('.jpeg', '') for _ in df['source_file_path']]
    # print(source)
    target=[_.replace('\\', '/').split('/')[-1].replace('.jpeg', '') for _ in df['target_file_path']]
    # print(target)
    similarities=[_.replace("[", '').replace("]", '') for _ in df['similarities']]
    # print(df['similarities'])
    # print(df['similarities'][231])
    # print(df['response'][231])
    # print(df['similarities'][232])
    # print(df['response'][232])
    # df.columns
    _list=[]
    for _ in similarities:
        if _=='':
            _list.append(0.0)
        else:
            _list.append(float(_.replace("'", '')))
    similarities=_list
    similarities
    df_result=pd.DataFrame([])
    df_result['source']=source
    df_result['target']=target
    df_result['similarities']=similarities
    return df_result

df_attgan_sim_parsed=parse(df_attgan_sim)
df_fawkes_sim_parsed=parse(df_fawkes_sim)
df_combined_sim_parsed=parse(df_combined_sim)
```
