```python
list_path=glob.glob(ospj(r'C:\Users\Ranjani\Downloads\Sample_4.3\OpenFace\Input Images\Split_AttGAN', '*.png'))
list_f_name=[_.replace('\\', '/').split('/')[-1] for _ in list_path]
```

```python
path_split=r'C:\Users\Ranjani\Downloads\Sample_4.3\OpenFace\Input Images\Split_AttGAN'
engagement=np.concatenate([np.ones(15)*3, #000001.jpg_XXX.png
                        np.ones(15)*3, #000002.jpg_XXX.png
                        np.ones(15)*3, #000003.jpg_XXX.png
                        np.ones(15)*1, #000004.jpg_XXX.png
                        np.ones(15)*2  #000005.jpg_XXX.png
                        ], axis=0)
# print(len(list_f_name), len(engagement), len(list_path))
dummy=np.ones(75)*-1
clipid=np.concatenate([
    np.ones(15)*882654,
    np.ones(15)*510042,
    np.ones(15)*510038,
    np.ones(15)*510035,
    np.ones(15)*500044
], axis=0).astype(int)
df_label_split=pd.DataFrame(np.array([
    clipid, #ClipID
    dummy, #boredom
    engagement,
    dummy, #confusion
    dummy, #frustration
    list_f_name,
    list_path,
]).T, columns=['ClipID', 'Boredom', 'Engagement', 'Confusion', 'Frustration', 'f_name', 'path'])
```

```python
df_label_split.to_csv(ospj(path_split, 'label_split.csv'))
```
